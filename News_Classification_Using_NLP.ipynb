Pre Processing Steps

```python
dataset=dataset.reset_index(drop=True)
```

```python
dataset
```

```python
dataset=dataset.drop(['title','subject','date'],axis=1)
```

```python
dataset
```

```python
from nltk.tokenize import word_tokenize
dataset['text']=dataset['text'].apply(word_tokenize)
```

```python
dataset
```

```python
from nltk.stem.snowball import SnowballStemmer

```

```python
sb=SnowballStemmer('english')
```

```python
def stem_it(text):
    return[sb.stem(word) for word in text]
```

```python
dataset['text']=dataset['text'].apply(stem_it)
```

```python
dataset
```

```python
def stopword_removal(text):
    return [word for word in text if len(word) >>2]
```

```python
dataset['text']=dataset['text'].apply(stopword_removal)
```

```python
dataset
```

```python
dataset['text']=dataset['text'].apply(''.join)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(dataset['text'],dataset['Real'],test_size=0.25)
```

```python
x_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidfv=TfidfVectorizer(max_df=0.7)
```

```python
tfidfv_train=tfidfv.fit_transform(x_train)
tfidfv_test=tfidfv.transform(x_test)
```

```python
print(tfidfv_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=1000)

lr.fit(tfidfv_train,y_train)
```

```python
y_pred=lr.predict(tfidfv_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pac=PassiveAggressiveClassifier(max_iter=1000)
pac.fit(tfidfv_train,y_train)
```

```python
y_pred=pac.predict(tfidfv_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
